```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def cels2farh(cels):    
    cels = np.array(cels)
    return ((cels*9/5) + 32)
```

```python
day = "Mai"+str(2)
mois = "5Mai"
j1 = pd.read_csv(mois + "/" + day + ".csv")
j1 = j1.drop(columns=['Unnamed: 0'])
j1.plot(x='Time', y='OA T' )
OAT = j1['OA T']
```

```python
OAT = OAT.to_numpy()
OAT_f = cels2farh(OAT)
plt.plot(OAT_f)
```

```python
OAT = pd.DataFrame(data=OAT_f, columns=["OA T"])
j1['OA T'] = OAT
j1.plot(x='Time', y='OA T' )
```

```python
j1.to_csv(mois + "/" + day + ".csv")
```
