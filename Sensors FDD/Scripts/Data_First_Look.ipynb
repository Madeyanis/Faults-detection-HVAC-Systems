```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('MZVAV-1.csv')
data.columns
data = data[['AHU: Supply Air Temperature', 'AHU: Outdoor Air Temperature', 'AHU: Return Air Temperature', 'AHU: Mixed Air Temperature']]
data = data.rename(columns={'AHU: Supply Air Temperature': 'Supply Air Temp', 'AHU: Outdoor Air Temperature': 'Outdoor Air Temp', 'AHU: Return Air Temperature':'Return Air Temp', 'AHU: Mixed Air Temperature':'Mixed Air Temp'})
```

```python
boxprops = dict(linestyle='-', linewidth=4, color='b')
medianprops = dict(linestyle='-', linewidth=4, color='r')

data.boxplot(figsize=(10, 7), fontsize=13, boxprops=boxprops,
                medianprops=medianprops, showfliers=True, showmeans=False)
plt.savefig("Outliers_new", dpi='figure', format=None,
        bbox_inches=None, pad_inches=0.1,
        facecolor='auto', edgecolor='auto')  
```
